```python
from PIL import Image
import numpy as np
from sklearn import datasets
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def CosineSimilarity(i1, i2):
    return np.sum(i1*i2) /  ( np.sqrt( np.sum(i1**2)) * np.sqrt( np.sum(i2**2)) )
```

```python
def JaccardSimilarity(i1, i2):
    return len(list(set(i1)&set(i2))) / (len(list(set(i1)|set(i2))))
```

```python
def EuclidienneDistance(i1, i2):
    return np.sqrt(np.sum((i1-i2)**2))
```

Displaying an image from the handwritten digit database 

```python
digits = datasets.load_digits()
img=digits.images[3]
plt.imshow(img, cmap=plt.cm.gray_r, interpolation='nearest')
plt.show()
```

## Find the 10 similar images to the 2 handwrtitten image using  Euclidienne Distance

```python
image_desire=digits.images[2]
image_vect1=image_desire.flatten()
images_finale1 = []
for image in digits.images:
    #le remplissage d'une liste par toute les images de dataset avec leur distance 
    image_sim = {'image':image,'distance':EuclidienneDistance(image_vect1,image.flatten())}
    images_finale1.append(image_sim)

images_finale1.sort(key = lambda x:x['distance'])
plt.figure(figsize=(30,30))
plt.title(label="Result")
for i in range(20):
    image_and_desc = images_finale1[i]
    plt.subplot(5,10,i+1)
    plt.xticks([image_and_desc['distance']])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(image_and_desc['image'], cmap=plt.cm.binary)
plt.show()
# for res in images_finale:        
#     plt.imshow(res['image'], cmap=plt.cm.binary)
#     plt.show()
```

##Find the 10 similar images to the 2 handwrtitten image using  Jaccard Similarity

```python
image_desire=digits.images[0]
image_vect2=image_desire.flatten()
images_finale2 = []
for image in digits.images:
    #le remplissage d'une liste par toute les images de dataset avec leur distance 
    image_sim = {'image':image,'distance':JaccardSimilarity(image_vect2,image.flatten())}
    images_finale2.append(image_sim)
images_finale2.sort(reverse = True, key = lambda x:x['distance'])
plt.figure(figsize=(30,30))
plt.title(label="Result")
for i in range(20):
    image_and_desc = images_finale2[i]
    plt.subplot(5,10,i+1)
    plt.xticks([image_and_desc['distance']])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(image_and_desc['image'], cmap=plt.cm.binary)
plt.show()
# for res in images_finale:        
#     plt.imshow(res['image'], cmap=plt.cm.binary)
#     plt.show()
```
